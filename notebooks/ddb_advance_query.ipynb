## How to gen k bar data
```
group by
```

````python
ddb.run("""
k1min = select firstNot(Close) as Open,
               min(Close) as Low, 
               max(Close) as High, 
               lastNot(close) as Close, 
               sum(Volume) as Volume
               from tbMKT where Date >= 2020.05.01,
                                Date <= 2020.05.10,
                                Close > 0,
                                Time != 14:30:00 
                          group by Date, minute(Time)+1 as Time, code as Code
                          order by Code, Date, Time
k1min
""")
````

## How to apply some indicator to each symbol

```
context by
```

````python
ddb.run("""
select concatDateTime(Date, time(Time)) as DateTime,
      Code, Open, High, Low, Close, Volume,
      ta::sma(Close, 60) as ma60,
      ta::bBands(Close, 20, 2, 2, 0) as `upband`ma20`downband,
      natr(High, Low, Close, 20) as NATR
      from k1min context by Code
""").dropna()
````
